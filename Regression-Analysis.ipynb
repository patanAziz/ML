```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('lasso_ridge_dataset.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=data.corr(),annot=True)
plt.plot
plt.show()
```

```python
X = data.iloc[:,:-1]
y=data["target"]
```

```python
X
```

```python
y
```

```python
sc= StandardScaler()
sc.fit(X)
X
```

```python
new_data=pd.DataFrame(sc.transform(X),columns=X.columns)
new_data
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_data,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression , Lasso , Ridge
```

```python
re = LinearRegression()
ls = Lasso()
ri = Ridge()
```

```python
re.fit(x_train,y_train)
```

```python
re.score(x_test,y_test)*100
```

```python
ls.fit(x_train,y_train)
ls.score(x_test,y_test)*100
```

```python
ri.fit(x_train,y_train)
ri.score(x_test,y_test)*100
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

# Regression

```python
plt.figure(figsize=(15,5))
plt.bar(new_data.columns,re.coef_)
plt.xlabel("Columns")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(new_data.columns,ls.coef_)
plt.xlabel("Columns")
plt.show()
```

```python
print(mean_squared_error(y_test, re.predict(x_test)))
print(mean_absolute_error(y_test,re.predict(x_test)))
```

```python
plt.figure(figsize=(15,5))
plt.bar(new_data.columns,ri.coef_)
plt.xlabel("Columns")
plt.show()
```

```python
df = pd.DataFrame({"col_names":new_data.columns, "LinearRegression":re.coef_, "Lasso":ls.coef_, "Ridge":ri.coef_})
```
