```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data,columns=[data.feature_names])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Lables'] = data.target
```

```python
df.head()
```

```python
X = df.drop('Lables',axis=1)
```

```python
X.head()
```

```python
y = df['Lables']
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
Scaled_input_data=pd.DataFrame(ss.fit_transform(X),columns=[data.feature_names])
Scaled_input_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(Scaled_input_data,y,test_size=0.2,random_state=42)
```

```python
from keras.layers import Dense, BatchNormalization , Dropout
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.regularizers import L2
```

```python
Ann = Sequential()
```

```python
Ann.add(Dense(15,input_dim=30,activation='relu'))
Ann.add(Dense(8,activation='relu'))
Ann.add(Dense(1,activation='sigmoid'))
```

```python
Ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
Ann.fit(x_train,y_train,batch_size=100,epochs=50,callbacks=EarlyStopping())
```

```python
prd=Ann.predict(x_test)
```

```python
prd1=Ann.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prd_data1=[]
for i in prd1:
    if i[0]>0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
prd_data=[]
for i in prd:
    if i[0]>0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
prd_data
```

```python
accuracy_score(y_test,prd_data)*100
```
