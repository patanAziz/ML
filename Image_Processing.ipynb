DISPLAYING THE IMAGE USING PILLOW


```python
import PIL
import matplotlib.pyplot as plt
img = PIL.Image.open('/content/VIER PFOTEN_2019-03-15_001-2886x1999-1920x1330.jpg')
```

```python
plt.imshow(img)
```

RESIZE THE IMAGE USING PILLOW


```python
img = img.resize((200,200))
```

```python
plt.imshow(img)
```

COVERT ARRY TO NUMPY ARRAY


```python
import numpy as np
img_array = np.array(img)
```

```python
type(img_array)
```

```python
img_array
```

NORMALIZE THE IMAGE


```python
img_array = img_array/255.0
```

```python
img_array
```

```python
import tensorflow as tf
```

```python
tf_image = tf.convert_to_tensor(img_array)
```

```python
tf_image
```

```python
img_tensor = tf.expand_dims(tf_image, axis=0)
```
